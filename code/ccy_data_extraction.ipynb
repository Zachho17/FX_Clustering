```python
import refinitiv.data as rd
import datetime as dt
import pandas as pd

security = ['AUD=','NZD=','EUR=','GBP=','CAD=','CHF=','JPY=',
            'AUDNZD=R','EURAUD=R','GBPAUD=R','AUDCAD=R','AUDCHF=R','AUDJPY=R',
            'EURNZD=R','GBPNZD=R','NZDCAD=R','NZDCHF=R','NZDJPY=R',
            'EURGBP=R','EURCAD=R','EURCHF=R','EURJPY=R',
            'GBPCAD=R','GBPCHF=R','GBPJPY=R',
            'CADCHF=R','CADJPY=R','CHFJPY=R']

date_0 = '1996-01-01'
date_1 = '2023-12-31'

rd.open_session()
markets = rd.get_history(universe= security,
                         fields=['MID_PRICE'],
                     interval="1W", start=date_0, end=date_1
                    )

markets.head()
```
